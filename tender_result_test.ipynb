#### 3. Choose parameters for training and split data on test/train

```python
# choose data for model application
y = df["Won"]
X = df[["BID DATE", "STATE", "PROJECT LOCATION", "PROJECT DESCRIPTION", "BID", "Pending", "PROPOSAL", "VALIANT BID AMOUNT"]]
```

```python
X["STATE_NUM"] = 0
```

```python
for j in range(len(state)):
    for i, item in enumerate(range(len(X.STATE))):
        if X.STATE[i] == state.keys()[j]:
            X.loc[i, "STATE_NUM"] = state.values()[j]
```

```python
X[:2]
```

```python
lbin = LabelBinarizer()
S = lbin.fit_transform(X["STATE_NUM"])
```

```python
S[:10]
```

```python
S_df = pd.DataFrame(S)
```

```python
S_df[:3]
```

```python
X = pd.merge(X,S_df,left_index=True, right_index=True)
X = X.drop("STATE_NUM", 1)
```

```python
X[:3]
```

```python
X["WON_NUM"] = 0
```

```python
X2 = X[["PROJECT LOCATION", "WON_NUM"]]
X2["WON_NUM"] = y
```

```python
X2[:2]
```

```python
data = X2.groupby("PROJECT LOCATION", as_index=False)
data = data.aggregate(sum)
```

```python
data[:2]
```

```python
for j in range(len(data.WON_NUM)):
    for i, item in enumerate(range(len(X.STATE))):
        if X["PROJECT LOCATION"][i] == data["PROJECT LOCATION"][j]:
            X.loc[i, "WON_NUM"] = data["WON_NUM"][j]
```

```python
X[:3]
```

```python
# split on trainset and testset
X_train_all, X_test_all, y_train, y_test = train_test_split(X, y, random_state=0)
X_train = X_train_all[:, 4::]
X_test = X_test_all[:, 4::]
```

#### 5. Random forest.

```python
# fit RandomForest classifier on all integer features except "Lost"
forest = RandomForestClassifier()
forest = forest.fit(X_train, y_train)

# get prediction
pred_prob = forest.predict_proba(X_test)

# calculate AUC
pred = forest.predict(X_test).astype(int)
metrics.roc_auc_score(y_test, pred, average='macro', sample_weight=None)
```

```python
# choose best parameters for training RandomForest classifier
parameters = {'n_estimators':[100, 200, 500, 1000]}
forest = RandomForestClassifier()
forest = grid_search.GridSearchCV(forest, parameters)
forest.fit(X_train, y_train)
print "Best parameters by Search grid:", forest.best_params_
```

```python
# fit RandomForest classifier with {'n_estimators': 500}
forest = RandomForestClassifier(n_estimators = 500)
forest = forest.fit(X_train, y_train)

# get prediction
pred_prob = forest.predict_proba(X_test)

# calculate AUC
pred = forest.predict(X_test).astype(int)
metrics.roc_auc_score(y_test, pred, average='macro', sample_weight=None)
```

#### 7. Write results to file.

```python
probability = pd.DataFrame()
probability["Won"] = y_test
probability["Probability_of_success"] = 0
probability["Probability_of_success"] = pred_prob[:, 1]
probability.to_csv("Probability_of_success_historical.csv", index=False)
probability[:30]
```
