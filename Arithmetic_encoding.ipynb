```python
def data_load(path):
    eng_dict = open(path, 'r')
    eng_dict = eng_dict.read().split("\n")
    df = pd.DataFrame(eng_dict)
    return df

def letter_num(letters):
    # compute number of letters in dictionary
    for let in letters:
        letter_unique[let] = letter_unique.setdefault(let, 0) + 1              
    return letter_unique

def letter_number(dict):
    S = sum(dict.values())
    return S

def probability(dict):
    for letters in dict:
        dict[letters] = [dict[letters], (float(dict[letters])/let_number)]
    return dict
    

def mapping(d):
    a = 0
    sort = sorted(d.keys()) 
    for letters in sort:
        d[letters].append((a, a + d[letters][1]))
        a = a + d[letters][1]
    return d

def word_let(df):
    phonemic = True
    if " " in df[0]:
        word = dict()
        for item in df:
            # put to lower case and split  
            item = item.lower().split()
            
            # split the words
            words = item[0]
            
            # split the letters
            letters = item[2:len(item)]
            
            # add the '#'  the final symbol
            letters.append('#')
            
            # create corpus with words and letters
            word[words] = letters 
    else:
        phonemic = False
        word  = []
        for item in df:
            word.append(item + "#")        
    return word

def from_right(words):
    if words is list:
        words.reverse()
        w = words.pop(0)
        words.append(w)
    if words is str:
        words[::-1]
        words = words[1:] + words[1]
    return words

def corpus_creating(df):
    phonemic = True
    if " " in df[0]:
        corpus = dict()
        for item in df:
                # put to lower case and split  
                item = item.lower().split()
            
                # split the words
                words = item[0]
            
                # split the letters
                letters = item[2:len(item)]
            
                # add the '#'  the final symbol
                letters.append('#')
            
                # create corpus with words and letters
                corpus[words] = letters           
             
                # compute number of letters in dictionary
                letter_unique = letter_num(letters)
    else:
        phonemic = False
        corpus  = []
        for item in df:
            corpus.append(item + "#")
            letter_unique = letter_num(item)
         
    return letter_unique


def left_right_iteration(corpus, words):
    probability = []
    for i in range(2):
        (left, right) = corpus[words[i]][2]
        for word in words[1:]:
            (begin, end) = corpus[word][2]
            left  = left * (1 - begin) + (begin * right)
            right = left + (end - begin) * (right - left)
        probability.append((left, right))
        words = from_right(words)    
       
    return probability


def write_to_file(path, probability):
    with open(path, 'w') as text:
        for word in sorted(probability.keys()):
            text.write("{w} {b1} {e1} {b2} {e2}\n".format(w = word, 
               b1 = probability[word][0][0], e1 = probability[word][0][1], 
               b2 = probability[word][1][0], e2 = probability[word][1][1]))
            
def coordinates_x(probability):
    x = []
    for item in probability_in:
        x.append(probability_in[item][0][0])
        y.append(probability_in[item][1][0])
    return x
        
def coordinates_y(probability):
    y = []
    for item in probability_in:
        y.append(probability_in[item][1][0])
    return y
    
```

```python
path1 = "EnglishSmall.dx1.txt"
path2 = "result2.txt"
df = data_load(path1)
corpus = corpus_creating(df[0])
let_number = letter_number(corpus)
corpus = probability(corpus)
corpus = mapping(corpus)
words = word_let(df[0])
probability_in = {k: v for k,v in zip(words, [left_right_iteration(corpus, words[w]) for w in words])}
write_to_file(path2, probability_in) 
x = coordinates_x(probability_in)
y = coordinates_y(probability_in)
```
